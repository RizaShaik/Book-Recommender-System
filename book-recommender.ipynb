## Popularity based book recommender

```python
ratings_with_names = ratings.merge(books,on='ISBN')
```

```python
num_ratings = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_ratings.rename(columns={'Book-Rating' : 'num_ratings'},inplace = True)
num_ratings
```

```python
avg_ratings = ratings_with_names.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings.rename(columns={'Book-Rating' : 'avg_rating'},inplace = True)
avg_ratings
```

```python
popular_books = num_ratings.merge(avg_ratings,on='Book-Title')
popular_books
```

```python
popular_books = popular_books[popular_books['num_ratings']>=300].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_books = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_books
```

```python
popular_books['Image-URL-M'][0]
```

```python
import pickle
pickle.dump(popular_books,open('popular.pkl','wb'))
```

### Collaborative filtering

```python
temp_users = ratings_with_names.groupby('User-ID').count()['Book-Rating'] > 200
readers = temp_users[temp_users].index
readers
```

```python
filtered_rating = ratings_with_names[ratings_with_names['User-ID'].isin(readers)]
filtered_rating
```

```python
temp_books = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
popular_books = temp_books[temp_books].index
popular_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(popular_books)]
final_ratings.drop_duplicates()
```

```python
table = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
table
```

```python
table.fillna(0,inplace=True)
table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

To calculate Euclidean distance:

```python
similarity_score = cosine_similarity(table) 
similarity_score
```

```python
def recommend(book_name):
#index fetch
    index = np.where(table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:11]
    data = []
    for i in similar_items:
        items = []
        temp_books = books[books['Book-Title'] == table.index[i[0]]]
        items.extend(list(temp_books.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_books.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_books.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(items)
    return data
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```

```python
pickle.dump(table,open('table.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similarity_score.pkl','wb'))
```
